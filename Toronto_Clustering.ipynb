# **Segmenting and Clustering Neighborhoods in Toronto**


## Project Description
The aim of this porject is to cluster different neighborhoods in Toronto by k-means clustering machine learning algorithm which can be utilized by individuals or businesses. 

For the Toronto neighborhood data, a Wikipedia page exists that has all the information needed to explore and cluster the neighborhoods in Toronto. I will  scrape the Wikipedia page and wrangle the data, clean and then read it into a pandas dataframe.

```python
#Importing the necessary libraries and packages

import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup #Pulling data out of HTML and XML files
from geopy.geocoders import Nominatim
import folium
from IPython.display import display_html
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

```

### Wikipedia Scraping and Wrangling the Data

```python
#Getting the html file from Wikipedia page with BeautifulSoup library 
html_doc = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
soup = BeautifulSoup(html_doc, 'lxml') # Soup object to be parsed
table=str(soup.table)
display_html(table,raw=True) # visualize the table


```

### Preprocessing and Cleaning Data

```python
# Convert html table first to df
df=pd.read_html(str(table))[0]
df.head()
```

```python
#Dropping the rows with not assigned Borough
df.drop(df[ df['Borough'] == 'Not assigned' ].index, inplace=True)
df.head()
```

```python
#Grouping the neighborhoods with same postal code
df = df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(", ".join).reset_index()

#If neighborhood not assigned change it to Borough name
df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned',df['Borough'], df['Neighbourhood'])
df.shape
```

### Importing latitude and longitude coordinates for postal codes in Toronto

```python
df_coor=pd.read_csv('https://cocl.us/Geospatial_data')
df_coor.head()
```

```python
# Merging two dfs based on postal code column. Inner left join is used as merging method.
df_toronto=pd.merge(df,df_coor, how='left', left_on='Postal Code', right_on='Postal Code')
df_toronto
```

```python
# Reducing the dataset to boroughs that have 'Toronto' in its name
df_toronto= df_toronto[df_toronto['Borough'].str.contains('Toronto',regex=False)]
df_toronto.head()
```

### Visualizing the Neighborhoods on Map using Folium


```python
# Getting latitude and longitude values of Toronto
address = "Toronto, ON"

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto city are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
for lat,lng,borough,neighbourhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### Modeling with k-Means Clustering

```python
df_clustering = df_toronto.drop(['Postal Code','Borough','Neighbourhood'],axis=1)

#Choosing number of clusters 5 (random) and fitting the data into model
k=5
kMeans = KMeans(n_clusters = k,random_state=3).fit(df_clustering)
df_toronto['Cluster_Labels']=kMeans.labels_
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighbourhood'], df_toronto['Cluster_Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
